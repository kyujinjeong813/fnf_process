```python
import pandas as pd

file_path = 'C://Users/kyujin/Desktop/PROCESS/MKT/MLB_MKT/mlb_hat_influencer.xlsx'
sheet_name = 'Sheet0'

df_mkt = pd.read_excel(file_path, sheet_name=sheet_name, header=0)
df_mkt = df_mkt.iloc[:,:-3]
df_mkt
```

```python
df_mkt.columns
```

```python
cols = ['대표닉네임','팔로워 수', '성별', '분야', '구분','링크','업로드일', '업로드기한', '좋아요', '댓글수']
df_mkt_sub = df_mkt[cols]

bins = [0, 10000, 30000, 50000, 100000, 300000, 500000, 1000000, 2000000, 3000000, 100000000]
pd.cut(df_mkt_sub['팔로워 수'], bins=bins)
df_mkt_sub['팔로워 점수'] = pd.cut(df_mkt_sub['팔로워 수'], bins=bins, labels=[5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
df_mkt_sub
```

```python
df_mkt_sub['인터랙션'] = df_mkt_sub['좋아요']+df_mkt_sub['댓글수']
df_mkt_sub
```

```python
df_mkt_sub['업로드일'] = df_mkt_sub['업로드일'].astype('datetime64[ns]')
df_mkt_sub['업로드기한'] = df_mkt_sub['업로드기한'].astype('datetime64[ns]')
```

```python
df_mkt_sub['기한'] = df_mkt_sub['업로드일'] - df_mkt_sub['업로드기한']
```

```python
from datetime import timedelta, date
td_1 = timedelta(days=0)
#td_2 = timedelta()
(df_mkt_sub['기한'] > td_1).sum()

# 지금 상태에선 기한 초과한 건 아예 집계가 안됨
```

```python
df_mkt_sub['인터랙션_등급'] = df_mkt_sub['인터랙션'].rank(pct=True)*20
df_mkt_sub
```

```python
#df_mkt_sub['정량_1'] = df_mkt_sub['팔로워 점수'] + df_mkt_sub['인터랙션_등급']
#df_mkt_sub[['대표닉네임','정량_1']]

df_mkt_sub['팔로워 점수_1'] = pd.to_numeric(df_mkt_sub['팔로워 점수'])
```

```python
df_mkt_sub['정량_1'] = df_mkt_sub['팔로워 점수_1'] + df_mkt_sub['인터랙션_등급']
df_mkt_sub[['대표닉네임','정량_1']].sort_values(['정량_1'], ascending=[False])

df_mkt_final = df_mkt_sub[['대표닉네임', '성별', '분야', '구분', '링크', '좋아요', '댓글수','팔로워 수','팔로워 점수', '정량_1']].sort_values(['정량_1'], ascending=[False])
df_mkt_final.to_excel('hat_test1.xlsx')
```
