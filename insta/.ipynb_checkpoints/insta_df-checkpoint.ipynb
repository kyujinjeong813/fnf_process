```python
import numpy as np
import pandas as pd
import os

base_dir = 'C:/Users/kyujin/Documents/Python Scripts/'
ex_file = 'insta_crawling1.xlsx'

ex_dir = os.path.join(base_dir, ex_file)

df = pd.read_excel(ex_dir, header=None, dtype={'user_id': str, 'key_url':str, 'init_time':str,'like':np.int64, 'view':np.int64,'hashtag':str},
                     fill_na=0)
df.columns = ['user_id', 'key_url', 'init_time', 'like', 'view', 'hashtag']
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
user_list = df['user_id'].unique().tolist()
user_list
```

```python
mask = df.iloc[:, 5].str.contains('디스커버리', na=False, regex=True)
mask
```

```python
df['mkt'] = 0
df.loc[mask,'mkt'] = 1
```

```python
# 사진 기준 좋아요 수만 집계
df['like'].dropna(inplace=True)
df = df[df.like != 0]
```

```python
df[df.mkt==1]
```

```python
#이걸 데이터프레임으로 저장..할까.. 아냐 일단은 탐색먼저

for user in user_list:
    df_sub = df[df['user_id'] == user]
    like_mean = df_sub['like'].mean()
    like_min = df_sub['like'].min()
    like_max = df_sub['like'].max()
    print(user, like_mean, ' / ',like_min, ' / ', like_max, '\n')
```

```python
# user group을 나누자
grouped = df['like'].groupby(df['user_id'])
like_mean = grouped.mean()
```

```python
like_mean = pd.DataFrame(like_mean)
like_mean
```

```python
like_mean = like_mean.sort_values(['like'], ascending=[False])
like_mean
```

```python
tier_3 = pd.qcut(like_mean['like'], 3)
tier_3
```

```python
like_mean.groupby(tier_3).agg(['count', 'mean', 'std'])
```

```python
like_mean['tier'] = pd.qcut(like_mean['like'], 3, labels=['3', '2','1'])
```

```python
# user_id와 tier를 짝지어서 원래 데이터프레임에 반영하고싶음
user_tier_dic = like_mean['tier'].to_dict()
```

```python
user_tier_dic
```

```python
df['tier'] = df['user_id'].apply(lambda x : user_tier_dic[x])
df
```
