```python
import pandas as pd
import numpy as np
import datetime

df = pd.read_excel('DX_MKT_2020.xlsx', sheet_name='KEYWORD_DATA')
keywords = set(df['KEYWORD'])

df_list = []
for word in keywords:
    df_word = df[df['KEYWORD'] == word]
    df_list.append(df_word)

new_df_list = []
for i in range(len(df_list)):
    datetime_series = pd.to_datetime(df_list[i]['SEARCHDATE'])
    datetime_index = pd.DatetimeIndex(datetime_series.values)
    new_df = df_list[i].set_index(datetime_index)
    new_df.drop('SEARCHDATE', axis=1, inplace=True)
    new_df_daily = new_df.asfreq('D')
    new_df_daily = new_df_daily.fillna(method='ffill')
    new_df_list.append(new_df_daily)

# 주간단위로 변경
keyword_df = pd.concat(new_df_list)
keyword_df = keyword_df.groupby('KEYWORD').resample('W-Sun').mean()
keyword_df = keyword_df.reset_index()
keyword_df.columns = ['KEYWORD', 'WDATE', 'SEARCHCOUNT']
# 기준일자 일->월로 변경
w_series = pd.to_datetime(keyword_df['WDATE'])
w_series = w_series.apply(lambda x: x + datetime.timedelta(days=1))
keyword_df['WDATE'] = w_series

keyword_df.to_excel('dx_keyword.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
import datetime

df = pd.read_excel('C://Users//kyujin//Desktop//PROCESS//MKT//DX//DX_MKT_2020.xlsx', sheet_name='SALE_DATA')
df = df[['WDATE', 'PARTCODE', 'SALE_QTY', 'SALE_AMT']]

df['WDATE'] = df['WDATE'].apply(lambda x: str(x).split(' ')[0])
df['WDATE']
```

```python
datetime_series = pd.to_datetime(df['WDATE'])
datetime_series
datetime_index = pd.DatetimeIndex(df['WDATE'].values)
datetime_index
```

```python
sale_df = df.set_index(datetime_index)

sale_df_w = sale_df.groupby('PARTCODE').resample('W-Sun').sum()
sale_df_w = sale_df_w.reset_index()
```

```python
sale_df_w.columns = ['PARTCODE', 'START_DT', 'SALE_QTY', 'SALE_AMT']
sale_df_w
```

```python
sale_df_w.loc[sale_df_w['SALE_QTY']<0, 'SALE_QTY'] = 0
sale_df_w.loc[sale_df_w['SALE_AMT']<0, 'SALE_AMT'] = 0
sale_df_w
```

```python
w_series = pd.to_datetime(sale_df_w['START_DT'])
w_series = w_series.apply(lambda x: x + datetime.timedelta(days=1))
sale_df_w['WDATE'] = w_series

sale_df_w.to_excel('dx_sales2020.xlsx', index=False)
```

```python
df_mkt['총비용'] = df_mkt['진행비용(\)']+df_mkt['제품비용(\)']
df_mkt['댓글'] = df_mkt['댓글'].fillna(0)
df_mkt['좋아요'] = df_mkt['좋아요'].fillna(0)
df_mkt['공유,저장'] = df_mkt['공유,저장'].fillna(0)
df_mkt = df_mkt.replace({'댓글': '-'}, {'댓글': 0})

df_mkt['좋아요'] = pd.to_numeric(df_mkt['좋아요'])
df_mkt['댓글'] = pd.to_numeric(df_mkt['댓글'])
df_mkt['공유,저장'] = pd.to_numeric(df_mkt['공유,저장'])
```

```python
df_mkt['인터랙션'] = df_mkt['좋아요'] + df_mkt['댓글']*5 + df_mkt['공유,저장']
```
